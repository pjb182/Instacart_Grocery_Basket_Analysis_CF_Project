# 02. If-Statements with User-Defined Functions

```python
# Create a subset of ords_prods_merged to save memory

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Define a function

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return np.nan
```

```python
# Apply a function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# No High-range products, probably due to this being a subset
```

```python
# Use max() to find highest priced item in new subset

df['prices'].max()
```

```python
# max() function shows the highest priced item in the entire df is 14.8 (less than 15) so there are not High-range products
```

# 04. If-Statements with For-Loops

```python
# Use a for-loop to create a new column in ords_prods_merged that summarizes how busy each day of the week is
```

```python
# First, find busiest day with frequency

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# 0 = Sat., 1 = Sun. etc.
# Sat. is busiest, Wed. is slowest
```

```python
# Create for-loop 

result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# print result

result
```

```python
# Create new col in ords_prods_merged and set it = to result to combine new result list

ords_prods_merged['busiest_day'] = result
```

```python
# Check new col frequency with 'orders_day_of_week' for consistency

ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Frequencies are consistent
```

## Q.4 Create a new “busiest_period_of_day" column.

```python
# First, find busiest hours with frequency

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# To divide into 3 categories I will use mean() and std()
```

```python
# Average of value_counts for order_hour_of_day

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).mean()
```

```python
# Quartiles of value_counts

np.percentile(ords_prods_merged['order_hour_of_day'].value_counts(dropna = False), [25, 50, 75])
```

```python
#Std Dev. of value_counts

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).std()
```

```python
# > avg + std = Most orders, < avg - std = Fewest orders, all else = Average orders
```

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).mean() + ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).std()
```

```python
# 9-16 = Most
```

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).mean() - ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).std()
```

```python
# 0-5 = Fewest, 6-8,17-23 = Avg
```

```python
# Create for-loop 

result_v3 = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value >= 9 and value <= 16:
    result_v3.append("Most orders")
  elif value < 6:
    result_v3.append("Fewest orders")
  else:
    result_v3.append("Average orders")
```

```python
result_v3
```

```python
# Create new col in ords_prods_merged and set it = to result to combine new result list

ords_prods_merged['busiest_period_of_day'] = result_v3
```
