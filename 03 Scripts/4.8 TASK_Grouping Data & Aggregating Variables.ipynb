## 2. Find the aggregated mean of the “order_number” column grouped by “department_id” for entire df.

```python
# Split the data into groups based on “department_id.”
# Apply the agg() function to each group to obtain the mean values for the “order_number” column.

ords_prods_merge.groupby('department_id').agg({'number_of_orders': ['mean']})
```

## 3. Analyze the result. How do the results for the entire dataframe differ from those of the subset?

### In the subset, deparments 16 and 19 had the highest average, but when using the entire df, departments 21 and 10 sit at the top. In addition, the overall averages of each department have decreased with the inclusion of the other 31M records (save for dept. 17).

## 4. Create a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Split df into groups using user_id
# Apply transform() to number_of_orders col 
# New 'max_order' is now created
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['number_of_orders'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Check code execution by looking at only a limited number of necessary columns 

ords_prods_merge[['user_id', 'number_of_orders', 'max_order']].head(20)
```

```python
# Create a flag to assign loyalty label using loc()
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check output with value_counts()

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check code execution by looking at only a limited number of necessary columns 

ords_prods_merge[['user_id', 'loyalty_flag', 'number_of_orders']].head(60)
```

## 5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer).

```python
# Perfom multiple arguments by adding more arguments (mean, min, max) to find basic statistics of loyalty category
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### NEW CUSTOMERS average the highest spend per order at 13.29, while LOYAL CUSTOMERS average the lowest spend per order at 10.39 and REGULAR CUSTOMERS average 12.50 per order.

## 6. Create a spending flag for each user based on the average price across all their orders.

```python
# Find the average price per item for each user and create new column - 'avg_item_price'

ords_prods_merge['avg_item_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
# Create a flag to assign spending label using loc()
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_item_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_item_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check output with value_counts()

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check code execution by looking at only a limited number of necessary columns 

ords_prods_merge[['user_id', 'avg_item_price', 'spending_flag']].head(60)
```

## 7. Create an order frequency flag that marks the regularity of a user’s ordering behavior.

```python
# Find the 'days_since_prior_order' median for each user and create new column - 'prior_order_median'

ords_prods_merge['prior_order_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# Create a flag to assign order_frequency label using loc()
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prior_order_median'] > 10) & (ords_prods_merge['prior_order_median'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['prior_order_median'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.head(15)
```

```python
# Check output with value_counts()

ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```

```python
# Check code execution by looking at only a limited number of necessary columns 

ords_prods_merge[['user_id', 'prior_order_median', 'order_freq_flag']].head(60)
```
